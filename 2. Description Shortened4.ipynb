```python
filename = 'Central and Western District.txt'

with open(filename, 'r', encoding="utf-8", errors='ignore') as f:
  text = f.read()

text = text.replace('Facilities accessible to wheelchair users but not up to local design standard', 'mobility')
text = text.replace('Facilities accessible to wheelchair users and meeting local design standard', 'mobility')
text = text.replace('Car park with reserved parking space for disabled drivers', 'mobility (parking)')
text = text.replace('Facilities available for hearing impaired persons', 'hearing')
text = text.replace('Facilities available for visually impaired persons', 'visual')
text = text.replace('Facilities inaccessible to wheelchair users', '')
text = text.replace('Locked', 'locked facility')
text = text.replace('Parking facilities available', '')
text = text.replace('Facilities unavailable', '')
text = text.replace('Image of', '')

with open(filename, 'w', encoding="utf-8", errors='ignore') as f:
  f.write(text)
```
